# ***Apply multinomial logistic regression Model***

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris(as_frame=True)
```

```python
data.data
```

```python
data.target
```

```python
df=data.frame
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['target'].unique()
```

```python
y = df['target']
```

```python
y
```

```python
x = df.drop('target', axis=1)
```

```python
x
```

```python
train_x, test_x, train_y, test_y=train_test_split(x, y, test_size=0.1, random_state=42 )
```

```python
mlr = LogisticRegression(max_iter=500, multi_class= 'multinomial')
```

```python
mlr.fit(train_x, train_y)
```

```python
model_train_pred = mlr.predict(train_x)
```

```python
print(classification_report(model_train_pred, train_y))
```

```python
model_test_pred = mlr.predict(test_x)
```

```python
print(classification_report(model_test_pred, test_y))
```
